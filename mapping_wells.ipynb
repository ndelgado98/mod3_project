```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
tan_map = gpd.read_file('TAN-level_1_SHP/TAN-level_1.shp')
```

```python
fig, ax = plt.subplots(figsize=(10,10))
tan_map.plot(ax=ax)
```

```python
df = pd.read_csv('df_clean.csv')
crs = {'init': 'espg:4326'}
```

```python
geometry = [Point(xy) for xy in zip(df['longitude'],df['latitude'])]
```

```python
geo_df = gpd.GeoDataFrame(df,crs=crs,geometry=geometry)
geo_df['geometry'].is_empty.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12,12))
tan_map.plot(ax=ax, alpha=0.5, color='gray')
geo_df[geo_df['status_group']=='functional'].plot(ax=ax, markersize=2, color='seagreen', label='Functional')
geo_df[geo_df['status_group']=='functional needs repair'].plot(ax=ax, markersize=2, color='goldenrod', label='Functional needs repair')
geo_df[geo_df['status_group']=='non functional'].plot(ax=ax, markersize=2, alpha=0.5,color='maroon', label='Non functional')
plt.xlim(29,41)
plt.ylim(-12,-0.5)
plt.legend(loc=1)
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```
